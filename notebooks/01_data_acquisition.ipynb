```python
import pandas as pd
import requests
import os
import sys

sys.path.append(os.path.abspath("../src"))

from config import *

print("Environment:", ENVIRONMENT)
print("Base URL loaded:", JSEARCH_BASE_URL)
print("Host loaded:", JSEARCH_HOST)
print("API key loaded:", JSEARCH_API_KEY is not None)
```

```python
url = f"{JSEARCH_BASE_URL}/search-v2"

querystring = {
    "query": "data analyst jobs in ontario",
    "page": "1",
    "num_pages": "1",
    "country": "ca",
    "date_posted": "all"
}

headers = {
    "x-rapidapi-key": JSEARCH_API_KEY,
    "x-rapidapi-host": JSEARCH_HOST
}

response = requests.get(
    url,
    headers=headers,
    params=querystring
)

print("Status Code:", response.status_code)
```
